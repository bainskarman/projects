```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly
```

```python
df=pd.read_csv('/workspaces/projects/credit_score/final.csv')
```

```python
df.drop(['Home Equity Loan','Auto Loan','Mortgage Loan','Payday Loan','Student Loan','Debt Consolidation Loan','Personal Loan','Credit-Builder Loan'],axis=1,inplace=True)
```

```python
df.drop(['Credit_Mix','Payment_of_Min_Amount','Num_Credit_Inquiries','Changed_Credit_Limit','Delay_from_due_date','Num_Bank_Accounts','Annual_Income','Monthly_Balance'],axis=1,inplace=True)
```

```python
df.head().T
```

```python
df.info()
```

```python
df= df[df['Payment_Behaviour'].notna()]
```

```python
df.head()
```

```python
df.describe()
```

```python
df['Num_of_Delayed_Payment'] = df['Num_of_Delayed_Payment'].apply(lambda x: max(0, x))
```

```python
df.describe()
```

```python
df.info()
```

```python
df['Credit_History_Age']=df['Credit_History_Age']//12
```

```python
df= df[['Occupation','Monthly_Inhand_Salary','Num_Credit_Card','Num_of_Loan','Num_of_Delayed_Payment','Outstanding_Debt','Credit_Utilization_Ratio','Credit_History_Age','Credit_Score']]
df
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
transformer= ColumnTransformer([('cat', OneHotEncoder(drop='first'),['Occupation'])],remainder="passthrough")
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop('Credit_Score',axis=1),df['Credit_Score'],test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

```python
pipeline = Pipeline([
    ('transformer', transformer),
    ('model', RandomForestClassifier())])
pipeline.fit(X_train,y_train)
pred=pipeline.predict(X_test)
accuracy_score(y_test,pred)
```
